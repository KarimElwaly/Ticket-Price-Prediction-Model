# Objective: 
The objective of this model is to predict flight prices given the various parameters. Data used in this model is publicly available at Kaggle. This will be a regression problem since the target is the price.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from flask import Flask, jsonify
from flask_restful import Api, Resource, reqparse
from flask_cors import CORS
from flask import Flask, Blueprint
from flask_cors import CORS
from flask_restful import Resource, Api
import config

sns.set()
```

```python
# Api
app = Flask(__name__)
CORS(app)
api = Api(app)

class FlightPrice(Resource):

    def post(self):
        parser = reqparse.RequestParser()

        parser.add_argument('Airline', type = list)                             
        parser.add_argument('Date_of_Journey', type = list)                            
        parser.add_argument('Source', type = list)                            
        parser.add_argument('Destination', type = list)
        parser.add_argument('Route', type = list)                             
        parser.add_argument('Dep_Time', type = list)                           
        parser.add_argument('Arrival_Time', type = list)                            
        parser.add_argument('Duration', type = list)
        parser.add_argument('Total_Stops', type = list)                            
        parser.add_argument('Additional_Info', type = list)                            
        parser.add_argument('Price', type = list)
        
        parser.add_argument('Airline1', type = list)                            
        parser.add_argument('Date_of_Journey1', type = list)                           
        parser.add_argument('Source1', type = list)                            
        parser.add_argument('Destination1', type = list)
        parser.add_argument('Route1', type = list)                            
        parser.add_argument('Dep_Time1', type = list)                            
        parser.add_argument('Arrival_Time1', type = list)                            
        parser.add_argument('Duration1', type = list)
        parser.add_argument('Total_Stops1', type = list)                            
        parser.add_argument('Additional_Info1', type = list)                          
        args = parser.parse_args()
        print(args)

        Airline = args['Airline']
        Date_of_Journey = args['Date_of_Journey']
        Source = args['Source']
        Destination = args['Destination']
        Route = args['Route']
        Dep_Time = args['Dep_Time']
        Arrival_Time = args['Arrival_Time']
        Duration = args['Duration']
        Total_Stops = args['Total_Stops']
        Additional_Info = args['Additional_Info']
        Price = args['Price']

        Airline1 = args['Airline1']
        Date_of_Journey1 = args['Date_of_Journey1']
        Source1 = args['Source1']
        Destination1 = args['Destination1']
        Route1 = args['Route1']
        Dep_Time1 = args['Dep_Time1']
        Arrival_Time1 = args['Arrival_Time1']
        Duration1 = args['Duration1']
        Total_Stops1 = args['Total_Stops1']
        Additional_Info1 = args['Additional_Info1']


        train_data = pd.DataFrame({'Airline' : Airline, 'Date_of_Journey' : Date_of_Journey, 'Source' : Source, 'Destination' : Destination, 'Route' : Route, 'Dep_Time' : Dep_Time, 'Arrival_Time' : Arrival_Time, 'Duration' : Duration, 'Total_Stops' : Total_Stops, 'Additional_Info' : Additional_Info, 'Price' : Price})


        test_data = pd.DataFrame({'Airline' : Airline1, 'Date_of_Journey' : Date_of_Journey1, 'Source' : Source1, 'Destination' : Destination1, 'Route' : Route1, 'Dep_Time' : Dep_Time1, 'Arrival_Time' : Arrival_Time1, 'Duration' : Duration1, 'Total_Stops' : Total_Stops1, 'Additional_Info' : Additional_Info1})

        mix = train_data.append(test_data)


        mix["Journey_day"] = pd.to_datetime(mix.Date_of_Journey, format="%d/%m/%Y").dt.day
        mix["Journey_month"] = pd.to_datetime(mix["Date_of_Journey"], format = "%d/%m/%Y").dt.month

        # Since we have converted Date_of_Journey column into integers, Now we can drop as it is of no use.
        mix.drop(["Date_of_Journey"], axis = 1, inplace = True)

        # Departure time is when a plane leaves the gate. 
        # Similar to Date_of_Journey we can extract values from Dep_Time

        # Extracting Hours
        mix["Dep_hour"] = pd.to_datetime(mix["Dep_Time"]).dt.hour

        # Extracting Minutes
        mix["Dep_min"] = pd.to_datetime(mix["Dep_Time"]).dt.minute

        # Now we can drop Dep_Time as it is of no use
        mix.drop(["Dep_Time"], axis = 1, inplace = True)

        # Arrival time is when the plane pulls up to the gate.
        # Similar to Date_of_Journey we can extract values from Arrival_Time

        # Extracting Hours
        mix["Arrival_hour"] = pd.to_datetime(mix.Arrival_Time).dt.hour

        # Extracting Minutes
        mix["Arrival_min"] = pd.to_datetime(mix.Arrival_Time).dt.minute

        # Now we can drop Arrival_Time as it is of no use
        mix.drop(["Arrival_Time"], axis = 1, inplace = True)

        # Time taken by plane to reach destination is called Duration
        # It is the differnce betwwen Departure Time and Arrival time


        # Assigning and converting Duration column into list
        duration = list(mix["Duration"])

        for i in range(len(duration)):
            if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
                if "h" in duration[i]:
                    duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
                else:
                    duration[i] = "0h " + duration[i]           # Adds 0 hour

        duration_hours = []
        duration_mins = []
        for i in range(len(duration)):
            duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
            duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration

        # Adding duration_hours and duration_mins list to train_data dataframe

        mix["Duration_hours"] = duration_hours
        mix["Duration_mins"] = duration_mins

        mix.drop(["Duration"], axis = 1, inplace = True)

        # As this is case of Ordinal Categorical type we perform LabelEncoder
        # Here Values are assigned with corresponding keys

        mix.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)

        train_data = mix[:len(train_data)]

        # From graph we can see that Jet Airways Business have the highest Price.
        # Apart from the first Airline almost all are having similar median

        # Airline vs Price
        sns.catplot(y = "Price", x = "Airline", data = train_data.sort_values("Price", ascending = False), kind="boxen", height = 6, aspect = 3)
        plt.show()

        # As Airline is Nominal Categorical data we will perform OneHotEncoding

        Airline = mix[["Airline"]]

        Airline = pd.get_dummies(Airline, drop_first= True)

        # Source vs Price

        sns.catplot(y = "Price", x = "Source", data = train_data.sort_values("Price", ascending = False), kind="boxen", height = 4, aspect = 3)
        plt.show()

        # As Source is Nominal Categorical data we will perform OneHotEncoding

        Source = mix[["Source"]]

        Source = pd.get_dummies(Source, drop_first= True)

        # As Destination is Nominal Categorical data we will perform OneHotEncoding

        Destination = mix[["Destination"]]

        Destination = pd.get_dummies(Destination, drop_first = True)

        # Additional_Info contains almost 80% no_info
        # Route and Total_Stops are related to each other

        mix.drop(["Route", "Additional_Info"], axis = 1, inplace = True)


        # Concatenate dataframe --> train_data + Airline + Source + Destination

        mix = pd.concat([mix, Airline, Source, Destination], axis = 1)

        mix.drop(["Airline", "Source", "Destination"], axis = 1, inplace = True)


        data_train = mix[:len(train_data)]

        test_data = mix[len(train_data):]
        test_data = test_data.loc[:, test_data.columns != 'Price' ]

        data_train.dropna(inplace = True)

        X = data_train.loc[:, data_train.columns != 'Price' ]

        y = data_train.iloc[:, data_train.columns == 'Price']


        # Finds correlation between Independent and dependent attributes

        plt.figure(figsize = (18,18))
        sns.heatmap(train_data.corr(), annot = True, cmap = "RdYlGn")

        plt.show()


        # Important feature using ExtraTreesRegressor

        from sklearn.ensemble import ExtraTreesRegressor
        selection = ExtraTreesRegressor()
        selection.fit(X, y)



        from sklearn.model_selection import train_test_split
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 32)


        from sklearn.ensemble import RandomForestRegressor
        reg_rf = RandomForestRegressor()
        reg_rf.fit(X_train, y_train)

        y_pred = reg_rf.predict(X_test)

        reg_rf.score(X_train, y_train)
        reg_rf.score(X_test, y_test)


        from sklearn import metrics
        from sklearn.model_selection import RandomizedSearchCV

        #Randomized Search CV

        # Number of trees in random forest
        n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
        # Number of features to consider at every split
        max_features = ['auto', 'sqrt']
        # Maximum number of levels in tree
        max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
        # Minimum number of samples required to split a node
        min_samples_split = [2, 5, 10, 15, 100]
        # Minimum number of samples required at each leaf node
        min_samples_leaf = [1, 2, 5, 10]

        # Create the random grid

        random_grid = {'n_estimators': n_estimators,
                       'max_features': max_features,
                       'max_depth': max_depth,
                       'min_samples_split': min_samples_split,
                       'min_samples_leaf': min_samples_leaf}

        # Random search of parameters, using 5 fold cross validation, 
        # search across 100 different combinations
        rf_random = RandomizedSearchCV(estimator = reg_rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=32, n_jobs = -1)

        rf_random.fit(X_train,y_train)

        rf_random.best_params_

        prediction = rf_random.predict(X_test)
        
        Accuracy = metrics.r2_score(y_test, prediction)
        
        
        solution = list(rf_random.predict(test_data))

        print({'solution': solution, 'Accuracy':Accuracy})
        return({'solution': solution, 'Accuracy':Accuracy})

    
api.add_resource(FlightPrice, '/FlightPrice')


   
if __name__ == '__main__':
    app.run()
```
